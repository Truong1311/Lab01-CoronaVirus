```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('2021-04-22_copy.csv', index_col=0)
```

```python
data
```

```python
fig, ax = plt.subplots(1,2,figsize = (18,6))
width = 0.35
# Tổng số ca nhiễm và tử vong của từng vùng( Continent)
ttsum_continent_cases = data.groupby(['Continent'])['TotalCases'].sum()
ttsum_continent_deaths = data.groupby(['Continent'])['TotalDeaths'].sum()

labels = ttsum_continent_cases.index
x = np.arange(len(labels))

ax[0].bar(x - width/2, ttsum_continent_cases.values,width,label = 'Cases')
ax[0].bar(x + width/2, ttsum_continent_deaths.values,width, label = 'Deaths')

ax[0].set_xticklabels(labels, rotation = 90)
ax[0].set_ylabel('Counts')
ax[0].set_title('Total Cases and Total Deaths in Continents')
ax[0].set_xticks(x)

#Số ca nhiễm và tử vong của từng vùng( Continent) trong ngày 22/4/2021
nwsum_continent_cases = data.groupby(['Continent'])['NewCases'].sum()
nwsum_continent_deaths = data.groupby(['Continent'])['NewDeaths'].sum()

ax[1].bar(x - width/2, nwsum_continent_cases.values,width,label = 'Cases')
ax[1].bar(x + width/2, nwsum_continent_deaths.values,width, label = 'Deaths')

ax[1].set_xticklabels(labels, rotation = 90)
ax[1].set_ylabel('Counts')
ax[1].set_title('New Cases and New Deaths in Continents')
ax[1].set_xticks(x)

plt.legend(bbox_to_anchor=(1.2,1))
plt.show()
```
