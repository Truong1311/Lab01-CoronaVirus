```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('2021-04-22_copy.csv', index_col=0)
```

```python
data
```

```python
#tính tỉ lệ % recovered và death
data['ratio_recovered']=data['TotalRecovered']/data['TotalCases']*100
data['ratio_death']=data['TotalDeaths']/data['TotalCases']*100
data
```

```python
#dùng top 10 nước có TotalCases cao nhất để trực quan hóa dữ liệu
top10_data=data.iloc[0:10]
top10_data
```

```python
#biểu đồ thể hiện % recovered và death của top 10 nước có TotalCases cao nhất 
plt.figure(figsize=(14,6))
countrys=list(top10_data['Country,Other'])
index = np.arange(10)
width = 0.5
plt.bar(index+width, top10_data['ratio_recovered'], color='Green',label='âsasas')
plt.bar(index+width, top10_data['ratio_death'], width, color='Red')
plt.title("(%)Recovery vs Deaths")
plt.ylabel("(%)")
plt.xticks(index+width, countrys)
```

### UK có tỉ lệ Recovered cao nhất, và tỉ lệ Death cũng tương đối cao, điều này có vẻ mâu thuẫn
### France có tỉ lệ Recovered thấp nhất và tỉ lệ Death cũng ở tỉ lệ trung bình so với các nước khác

```python
sns.lmplot(x="ratio_death", y="ratio_recovered", data=data)
plt.show()
```

### Có vài nước bất thường có tỉ lệ Recovered và Death rất cao

```python
#biểu đồ thể hiện tỉ lệ totalcases của top 10 nước có totaldeath cao nhất
labels=list(top10_data['Country,Other'])
plt.pie(top10_data['TotalCases'],labels=labels,autopct='%0.1f%%', startangle=-60)
plt.title("TotalCases %")
plt.axis('equal')
plt.legend(title="Countries")
plt.show()
```

```python
#biểu đồ thể hiện tỉ lệ totaldeaths của top 10 nước có totaldeath cao nhất
labels=list(top10_data['Country,Other'])
plt.pie(top10_data['TotalDeaths'],labels=labels,autopct='%0.1f%%', startangle=-60)
plt.title("TotalDeaths %")
plt.axis('equal')
plt.legend(title="Countries")
plt.show()
```

```python
#biểu đồ thể hiện tỉ lệ totalrecovered của top 10 nước có totaldeath cao nhất
labels=list(top10_data['Country,Other'])
plt.pie(top10_data['TotalRecovered'],labels=labels,autopct='%0.1f%%',startangle=-60)
plt.title("TotalRecovered %")
plt.axis('equal')
plt.legend(title="Countries")
plt.show()
```

```python
#biểu đồ thể hiện tỉ lệ Actioncases của top 10 nước có totaldeath cao nhất
labels=list(top10_data['Country,Other'])
plt.pie(top10_data['ActiveCases'],labels=labels,autopct='%0.1f%%',startangle=-60)
plt.title("ActiveCases %")
plt.axis('equal')
plt.legend(title="Countries")
plt.show()
```
