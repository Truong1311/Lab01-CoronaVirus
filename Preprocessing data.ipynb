```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
data = pd.read_csv('2021-04-22.csv')
```

```python
data = data.drop(['Unnamed: 0', '#'],axis=1)
Toltal = data.iloc[0]
data = data.drop(0, axis =0)
data = data.rename(columns={'Tests/\n1M pop\n':'Tests/1M pop'},inplace= False)
```

```python
cate_cols = ['Country,Other','Continent']
numberic_cols = data.columns.difference(cate_cols)
```

```python
numberic_cols
```

```python
data[numberic_cols]
```

```python
data[numberic_cols] = data[numberic_cols].apply(lambda x: x.str.replace(re.compile(r'[+.,]'), '').replace(' ',np.nan)).fillna('0').astype('int64')
```

- Quan sát Diamond Princess và MS Zaandam là hai du thuyền bị nhiễm corona nên sẽ fillna ở thuộc tính Continent là Ship

```python
data[cate_cols] = data[cate_cols].fillna('Ship')
```
